```python
from sklearn import datasets, linear_model
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
from sklearn.metrics import mean_squared_error
```

```python
df_new_train = cleaned_data.replace({False:0,True:1})
```

```python
X = df_new_train.drop(["title","author", "over_18"], axis = 1, inplace = False)
y = pd.DataFrame(df_new_train["over_18"])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state = 1)
```

```python
lm = linear_model.LinearRegression()
model = lm.fit(X_train, y_train)
predictions = lm.predict(X_test)
```

```python
print ('Test RMSE over_18:', np.sqrt(mean_squared_error(predictions, y_test)))
print ('Train RMSE over_18:', np.sqrt(mean_squared_error(lm.predict(X_train), y_train)))
```
